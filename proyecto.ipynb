```python
df = df_inicial[df_inicial['Year'] >= 2010]  # Filtrando el dataset únicamente para los años entre 2010 y 2020
atletas = df['Name'].unique()



deportes = list()
ranking2010 = list()
ranking2011 = list()
ranking2012 = list()
ranking2013 = list()
ranking2014 = list()
ranking2015 = list()
ranking2016 = list()
ranking2017 = list()
ranking2018 = list()
ranking2019 = list()
ranking2020 = list()


for atleta in atletas:
    #Deporte
    
    d =df[(df['Name']==atleta)]
    
    d = d.iloc[0].values
    deporte = d[5] #Deporte
    deportes.append(deporte) #LLena la lista deportes de acuerdo al atleta del ciclo for
    
    ## 2010
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2010)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2010.append(ranking)
    except:
        ranking = '-'
        ranking2010.append(ranking)

    ## 2011
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2011)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2011.append(ranking)
    except:
        ranking = '-'
        ranking2011.append(ranking)

    ## 2012
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2012)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2012.append(ranking)
    except:
        ranking = '-'
        ranking2012.append(ranking)


    ## 2013
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2013)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2013.append(ranking)
    except:
        ranking = '-'
        ranking2013.append(ranking)

    ## 2014
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2014)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2014.append(ranking)
    except:
        ranking = '-'
        ranking2014.append(ranking)

    ## 2015
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2015)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2015.append(ranking)
    except:
        ranking = '-'
        ranking2015.append(ranking)

    ## 2016
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2016)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2016.append(ranking)
    except:
        ranking = '-'
        ranking2016.append(ranking)
    
    ## 2017
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2017)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2017.append(ranking)
    except:
        ranking = '-'
        ranking2017.append(ranking)

    ## 2018
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2018)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2018.append(ranking)
    except:
        ranking = '-'
        ranking2018.append(ranking)

    ## 2019
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2019)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2019.append(ranking)
    except:
        ranking = '-'
        ranking2019.append(ranking)

    ## 2020
    try:
        r =df[(df['Name']==atleta) & (df['Year']==2020)]
        r = r.iloc[0].values
        ranking = r[3]
        ranking2020.append(ranking)
    except:
        ranking = '-'
        ranking2020.append(ranking)


df_aux = pd.DataFrame({'Nombre': atletas,
                        'Deporte': deportes,
                        '2010': ranking2010,
                        '2011': ranking2011,
                        '2012': ranking2012,
                        '2013': ranking2013,
                        '2014': ranking2014,
                        '2015': ranking2015,
                        '2016': ranking2016,
                        '2017': ranking2017,
                        '2018': ranking2018,
                        '2019': ranking2019,
                        '2020': ranking2020})



df_aux.to_csv('data/atletas_info_aux.csv', index=False)  

df_aux = pd.read_csv('data/atletas_info_aux.csv')
 

## Este códigpo es para mostrar el ranking máximo y mínimo de cada atleta
atletas = df_aux['Nombre'].values
atletas
max_ranking_list = list()
min_ranking_list = list()
#año_max_ranking_list = list()
#año_min_ranking_list = list()


for i in range(len(atletas)):
    atleta_info = df_aux.iloc[i,:]
    max_ranking = None
    min_ranking = None
    

    for i in range(len(atleta_info)):
        ranking = atleta_info[i]
        try:
            ranking = int(ranking)
            if max_ranking == None:
                max_ranking = ranking
            elif ranking < max_ranking:
                max_ranking = ranking 
                #año_max = atleta_info.index[i]

            if min_ranking == None:
                min_ranking = ranking
            elif ranking > min_ranking:
                min_ranking = ranking
                #año_min = atleta_info.index[i]
        except:
            pass
    max_ranking_list.append(max_ranking)
    min_ranking_list.append(min_ranking)
 
df_aux['Max_Ranking'] = max_ranking_list

df_aux['Min_Ranking'] = min_ranking_list

df_aux.to_csv('data/atletas_info_aux2.csv', index=False)

df_aux.head(35)
```

### Pregunta 2. Atleta con el menor número de apariciones y mayores ganancias.

```python
# Primero se suman las ganancias de todos los atletas
df_agrupado_ganancias = df_inicial.groupby(["Name"]).sum()
df_agrupado_ganancias.drop(['S.NO','Year','Current Rank'], axis=1,inplace=True)
df_agrupado_ganancias.reset_index(inplace=True)
df_agrupado_ganancias
```

```python
# Despúes se hace el recuento de cuantas veces apareció cada atleta
df_agrupado_apariciones = df_inicial.groupby(["Name"]).count()  #Se agrupa por nombre y se cuenta cuantas veces apareció
df_agrupado_apariciones.reset_index(inplace=True) 
df_agrupado_apariciones.drop(['Year','Current Rank','Nationality','Previous Year Rank','Sport','earnings ($ million)'], axis=1, inplace=True) #Se eliminan columnas no deseadas
df_agrupado_apariciones

```

```python
#Por último se combinan las dos tablas
df_menos_apariciones_mas_ingresos = pd.merge(df_agrupado_ganancias, df_agrupado_apariciones, on="Name") #Se une las dos tablas
df_menos_apariciones_mas_ingresos.rename(columns = {'S.NO':'Apariciones'}, inplace = True)
df_menos_apariciones_mas_ingresos = df_menos_apariciones_mas_ingresos[df_menos_apariciones_mas_ingresos['Apariciones'] == 1] #Se filtra por apariciones igual a 1
df_menos_apariciones_mas_ingresos.sort_values(by=['earnings ($ million)'], ascending=False, inplace=True) #Se ordena por ingresos
df_menos_apariciones_mas_ingresos.head()
```
